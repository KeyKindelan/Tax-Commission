<h1>Now let's explore entire dataset in chunks</h1>

```python
### analysis
## find all owners who show up more than 5 times presumable since they own multiple buildings
row_size = 500_000
count = 1
target_dfs = []

## this iterates thru huge dataset
for chunk in pd.read_csv("Property_Valuation_and_Assessment_Data_Tax_Classes_1_2_3_4_20250412.csv", chunksize=row_size,low_memory=False):
    print(f"Analyzing chunk {count}")
    count = count + 1

        #add owner_count column directly to the cnunk
    chunk["OWNER_COUNT"] = chunk.groupby("OWNER")["OWNER"].transform("count")

        #now query on the chunk
    filtered_chunk = chunk.query("OWNER_COUNT >= 10")

        #Don't change below this line
#_________________________________

    target_dfs.append(filtered_chunk)
print(f"Done processing {count} chunks")

final_df = pd.concat(target_dfs, ignore_index=True)
```

```python
#creating a new column with the tax rate. Trying on the sample_df first.

sample_df["tax_Rate"] = sample_df.apply(lambda x: '.20085' if x["CURTAXCLASS"] == "1"
                                else ('.12500' if x["CURTAXCLASS"] == "2"
                                else '.11181' if x["CURTAXCLASS"] == "3"
                                else '.10762' if x["CURTAXCLASS"] == "4"
                                else "0"), axis = 1)
sample_df
```

```python
#running the new column on the final_df
final_df["tax_Rate"] = final_df.apply(lambda x: '.20085' if x["CURTAXCLASS"] == "1"
                                else ('.12500' if x["CURTAXCLASS"] == "2"
                                else '.11181' if x["CURTAXCLASS"] == "3"
                                else '.10762' if x["CURTAXCLASS"] == "4"
                                else "0"), axis = 1)
final_df.sample(10)
```

```python
# Creating a taxed amount column, which is the taxable value minus the exemption x the tax rate. Testing on sample_df
sample_df["taxed_amt"] = sample_df.apply(lambda x: (float(x["CURTXBTOT"]) - float(x["CURTXBEXTOT"])) * float(x["tax_Rate"]), axis = 1)
sample_df
```

```python
# Running the same calculation on the final_df and looking at a sample of 10
final_df["taxed_amt"] = final_df.apply(lambda x: (float(x["CURTXBTOT"]) - float(x["CURTXBEXTOT"])) * float(x["tax_Rate"]), axis = 1)
final_df.sample(10)
```

```python
# I want to total the amount paid per boro. 
boro_sum = final_df.groupby("BORO")["taxed_amt"].sum()
boro_sum
```

```python
# reading the tax reduction data set to get total reduced per boro
reductions_df = pd.read_csv("tabula-2023 TC Actions2.csv")
reductions_df.sample(20)
```

```python
#summing reductions by boro
reductions_sum = reductions_df.groupby("BORO")["REDUCTION"].sum()
reductions_sum
```

```python
#creating a dataset that has total taxes paid and total deductions, then going to add a column for the proportion
boro_totals = pd.concat([boro_sum, reductions_sum], axis=1, join='outer')
boro_totals
```

```python
#Calculating which boro gets the biggest reductions 
boro_totals["BoroReduction_Per_TaxedAmt"] = boro_totals.apply(lambda x: (x["REDUCTION"]) / (x["taxed_amt"]), axis = 1)
boro_totals
```
